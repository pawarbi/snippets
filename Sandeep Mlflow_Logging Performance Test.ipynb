```python
import numpy as np
import mlflow
from itertools import product
from mlflow.tracking import MlflowClient
import pandas as pd
from dateutil import relativedelta

print("MLFlow version used: ", mlflow.__version__)
```

```python
##Defining params to sweep
param1 = np.linspace(1,10)
param2 = np.logspace(0,10)

## Function to create metric. This fn doesn't mean anything. 
def my_func(p1,p2):
  return np.sin(p1)+np.log(p2)
```

```python
EXP_NAME = "Sandeep_experiment1"
mlflow.set_experiment(experiment_name=EXP_NAME)
mlflow.start_run(run_name="Mlflow Logging Experiment")
```

```python
def run_exp():
  for param in product(param1, param2):
    with mlflow.start_run(run_name="Param_"+str(param[0])+"_"+str(param[1]), nested=True):
      mlflow.log_param("p1",param[0])
      mlflow.log_param("p2",param[1])
      mlflow.log_metric("result",my_func(param[0],param[1]))
      print(param[0], param[1], my_func(param[0],param[1]))
    
```

```python
run_exp()
```

```python
mlflow.end_run()
```

```python
runs = mlflow.search_runs(experiment_names=[EXP_NAME])
runs.head(3)
```

```python
duration = relativedelta.relativedelta( runs['start_time'].min(), runs['start_time'].max())
print(duration)
```
